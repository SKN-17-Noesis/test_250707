```python
import random
import time
import threading

# ğŸ§  ì„¤ì •ê°’
MAX_BUGS = 15           # ëŸ¬ë¸Œë²„ê·¸ ìµœëŒ€ ë“±ì¥ ìˆ˜
MAX_MISSES = 5          # ì¡°ê¸° ì¢…ë£Œ ê¸°ì¤€: 8ë§ˆë¦¬ ì´ìƒ ë“±ì¥ & 5ë§ˆë¦¬ ì´ìƒ ì‹¤íŒ¨
BUG_THRESHOLD = 8       # ì¡°ê¸° ì¢…ë£Œ íŒë‹¨ ê¸°ì¤€

# ğŸ“Š ìƒíƒœ ë³€ìˆ˜
bug_count = 0
defeated = 0
missed = 0
user_input_received = False

# ğŸ¯ ëŸ¬ë¸Œë²„ê·¸ ì²´ë ¥ ìƒì„±
def spawn_bug():
    hp = random.randint(5, 20)
    print(f"\nğŸ› ìƒˆë¡œìš´ ëŸ¬ë¸Œë²„ê·¸ ë“±ì¥! ì²´ë ¥: {hp}")
    return hp

# ğŸ’¥ í”Œë ˆì´ì–´ ê³µê²©
def player_attack():
    damage = random.randint(10, 30)
    print(f"ğŸ§¹ ê³µê²© ì‹œë„! ë°ë¯¸ì§€: {damage}")
    return damage

# ğŸ§¨ ì œí•œ ì‹œê°„ ì…ë ¥ (Windows í˜¸í™˜)
def wait_for_input():
    global user_input_received
    input()
    user_input_received = True

def timed_input(timeout_sec):
    global user_input_received
    user_input_received = False
    thread = threading.Thread(target=wait_for_input)
    thread.start()
    thread.join(timeout=timeout_sec)

    if not user_input_received:
        raise TimeoutError

# âš”ï¸ ëŸ¬ë¸Œë²„ê·¸ì™€ì˜ ì „íˆ¬ ì‹œë„
def attempt_to_defeat(bug_hp):
    print("â–¶ï¸ [Enter] í‚¤ë¥¼ ëˆŒëŸ¬ ê³µê²©í•˜ì„¸ìš”! (3ì´ˆ ì•ˆì—!)")
    try:
        timed_input(3)  # ì œí•œì‹œê°„ ì…ë ¥
    except TimeoutError:
        print("â° ì‹œê°„ ì´ˆê³¼! ë²Œë ˆê°€ ë„ë§ê°”ì–´ìš”!")
        return False

    damage = player_attack()
    if damage >= bug_hp:
        print("ğŸ¯ ëŸ¬ë¸Œë²„ê·¸ í‡´ì¹˜ ì„±ê³µ!")
        return True
    else:
        print("ğŸ˜± ë°ë¯¸ì§€ê°€ ë¶€ì¡±í•´! ë²Œë ˆê°€ ë‹¬ë¼ë¶™ì—ˆì–´ìš”...")
        return False

# ğŸ“Š í˜„ì¬ ìƒíƒœ ì¶œë ¥
def show_status(defeated, missed, total):
    print(f"\ní˜„ì¬ ìƒíƒœ: âœ… í‡´ì¹˜ {defeated} | âŒ ì‹¤íŒ¨ {missed} | ë“±ì¥í•œ ë²Œë ˆ ìˆ˜: {total}")

# ğŸ•¹ï¸ ê²Œì„ ì§„í–‰ ë£¨í”„
def lovebug_game():
    global bug_count, defeated, missed

    print("ğŸ’¥ ëŸ¬ë¸Œë²„ê·¸ í‡´ì¹˜ ì‘ì „ ì‹œì‘!")
    while bug_count < MAX_BUGS:
        bug_count += 1
        bug_hp = spawn_bug()

        success = attempt_to_defeat(bug_hp)
        if success:
            defeated += 1
        else:
            missed += 1

        show_status(defeated, missed, bug_count)

        # ì¡°ê¸° ì¢…ë£Œ ì¡°ê±´
        if bug_count >= BUG_THRESHOLD and missed >= MAX_MISSES:
            print("\nğŸš¨ ë„ˆë¬´ ë§ì€ ë²Œë ˆë¥¼ ë†“ì³¤ìŠµë‹ˆë‹¤! ê²Œì„ ì¢…ë£Œ...")
            break

        time.sleep(1)

    show_result(defeated, bug_count)

# ğŸ§¾ ìµœì¢… ê²°ê³¼ ì¶œë ¥
def show_result(defeated, total):
    print("\nğŸ® ê²Œì„ ì¢…ë£Œ!")
    rate = defeated / total
    print(f"\nìµœì¢… í‡´ì¹˜ìœ¨: {int(rate * 100)}%")

    if rate >= 0.8:
        print("âœ… ëŸ¬ë¸Œë²„ê·¸ í‡´ì¹˜ ì„±ê³µ! ë°©ì´ ê¹¨ë—í•´ì¡Œì–´ìš”! âœ¨")
    elif rate >= 0.4:
        print("âš ï¸ ì–´ë”˜ê°€ì— ìˆ¨ì–´ìˆì„ ëŸ¬ë¸Œë²„ê·¸ë¥¼ ì¡°ì‹¬í•´!")
    else:
        print("ğŸ¤¬ ë‹¹ì‹ ì˜ ì˜·ê³¼ ê°€ë°©ì„ ì¡°ì‹¬í•˜ì„¸ìš”... ëŸ¬ë¸Œë²„ê·¸ê°€ ë‚¨ì•„ìˆì–´ìš”!")

# ğŸ¬ ê²Œì„ ì‹œì‘
def start_menu():
    print("ğŸ ëŸ¬ë¸Œë²„ê·¸ í‡´ì¹˜ ê²Œì„ì— ì˜¤ì‹  ê±¸ í™˜ì˜í•©ë‹ˆë‹¤!")
    choice = input("ê²Œì„ì„ ì‹œì‘í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ").lower()
    if choice == "y":
        lovebug_game()
        
    else:
        print("ğŸ‘‹ ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ë‹¤ìŒì— ë‹¤ì‹œ ë„ì „í•´ì£¼ì„¸ìš”!")

# â–¶ï¸ ì‹¤í–‰
start_menu()
```

# 러브버그 퇴치 게임!

```python
import random
import time
import threading

# 🧠 설정값
MAX_BUGS = 15           # 러브버그 최대 등장 수
MAX_MISSES = 5          # 조기 종료 기준: 8마리 이상 등장 & 5마리 이상 실패
BUG_THRESHOLD = 8       # 조기 종료 판단 기준

# 📊 상태 변수
bug_count = 0
defeated = 0
missed = 0
user_input_received = False

# 🎯 러브버그 체력 생성
def spawn_bug():
    hp = random.randint(5, 20)
    print(f"\n🐛 새로운 러브버그 등장! 체력: {hp}")
    return hp

# 💥 플레이어 공격
def player_attack():
    damage = random.randint(10, 30)
    print(f"🧹 공격 시도! 데미지: {damage}")
    return damage

# 🧨 제한 시간 입력 (Windows 호환)
def wait_for_input():
    global user_input_received
    input()
    user_input_received = True

def timed_input(timeout_sec):
    global user_input_received
    user_input_received = False
    thread = threading.Thread(target=wait_for_input)
    thread.start()
    thread.join(timeout=timeout_sec)

    if not user_input_received:
        raise TimeoutError

# ⚔️ 러브버그와의 전투 시도
def attempt_to_defeat(bug_hp):
    print("▶️ [Enter] 키를 눌러 공격하세요! (3초 안에!)")
    try:
        timed_input(3)  # 제한시간 입력
    except TimeoutError:
        print("⏰ 시간 초과! 벌레가 도망갔어요!")
        return False

    damage = player_attack()
    if damage >= bug_hp:
        print("🎯 러브버그 퇴치 성공!")
        return True
    else:
        print("😱 데미지가 부족해! 벌레가 달라붙었어요...")
        return False

# 📊 현재 상태 출력
def show_status(defeated, missed, total):
    print(f"\n현재 상태: ✅ 퇴치 {defeated} | ❌ 실패 {missed} | 등장한 벌레 수: {total}")

# 🕹️ 게임 진행 루프
def lovebug_game():
    global bug_count, defeated, missed

    print("💥 러브버그 퇴치 작전 시작!")
    while bug_count < MAX_BUGS:
        bug_count += 1
        bug_hp = spawn_bug()

        success = attempt_to_defeat(bug_hp)
        if success:
            defeated += 1
        else:
            missed += 1

        show_status(defeated, missed, bug_count)

        # 조기 종료 조건
        if bug_count >= BUG_THRESHOLD and missed >= MAX_MISSES:
            print("\n🚨 너무 많은 벌레를 놓쳤습니다! 게임 종료...")
            break

        time.sleep(1)

    show_result(defeated, bug_count)

# 🧾 최종 결과 출력
def show_result(defeated, total):
    print("\n🎮 게임 종료!")
    rate = defeated / total
    print(f"\n최종 퇴치율: {int(rate * 100)}%")

    if rate >= 0.8:
        print("✅ 러브버그 퇴치 성공! 방이 깨끗해졌어요! ✨")
    elif rate >= 0.4:
        print("⚠️ 어딘가에 숨어있을 러브버그를 조심해!")
    else:
        print("🤬 당신의 옷과 가방을 조심하세요... 러브버그가 남아있어요!")

# 🎬 게임 시작
def start_menu():
    print("🐞 러브버그 퇴치 게임에 오신 걸 환영합니다!")
    choice = input("게임을 시작하시겠습니까? (y/n): ").lower()
    if choice == "y":
        lovebug_game()
        
    else:
        print("👋 게임을 종료합니다. 다음에 다시 도전해주세요!")

# ▶️ 실행
start_menu()
```
